## 1. Monthly Budget Snapshot

Ask the user for:
- monthly income
- rent
- food expenses
- transport expenses

Then:
1. compute total expenses
2. compute money left
3. compute % of income that was spent

Print a short report like:

```text
Total expenses: €X
Money left: €Y
You spent Z% of your income.

Round the percentage to 2 decimals.

````python
# Problem 1 – Monthly Budget Snapshot

monthly_income = float(input("Monthly income (€): "))
rent = float(input("Rent (€): "))
food_expenses = float(input("Food expenses (€): "))
transport_expenses = float(input("Transport expenses (€): "))

total_expenses = rent + food_expenses + transport_expenses
money_left = monthly_income - total_expenses
percent_spent = (total_expenses / monthly_income) * 100

print(f"Total expenses: €{total_expenses:.2f}")
print(f"Money left: €{money_left:.2f}")
print(f"You spent {percent_spent:.2f}% of your income.")
````

## 2. Two-Year GDP Change

Create two variables:
- `gdp_year_1`
- `gdp_year_2`

Compute:
- absolute change = new − old
- percentage change

Print:
```text
GDP increased by X billions, which is an increase of Y%.


````python
# Problem 2 – Two-Year GDP Change

gdp_year_1 = float(input("GDP in year 1 (in billions): "))
gdp_year_2 = float(input("GDP in year 2 (in billions): "))

absolute_change = gdp_year_2 - gdp_year_1
percentage_change = (absolute_change / gdp_year_1) * 100

print(f"GDP increased by {absolute_change:.2f} billions, which is an increase of {percentage_change:.2f}%.")
````

## 3. Compare Two Investment Options

Ask the user for the **final value** of:
- Investment A
- Investment B

Then:
1. print the higher final value (use `max(...)`)
2. print the lower final value (use `min(...)`)
3. print the difference between them

Example:
```text
Best investment returned: €1050
Worst investment returned: €1000
Difference: €50


````python
# Problem 3 – Compare Two Investment Options

investment_a = float(input("Final value of Investment A (€): "))
investment_b = float(input("Final value of Investment B (€): "))

best_investment = max(investment_a, investment_b)
worst_investment = min(investment_a, investment_b)
difference = best_investment - worst_investment

print(f"Best investment returned: €{best_investment:.2f}")
print(f"Worst investment returned: €{worst_investment:.2f}")
print(f"Difference: €{difference:.2f}")
````

## 4. Simple Salary Breakdown

Ask the user for:
- annual gross salary (€)
- tax rate (%)

Then compute:
- monthly gross salary
- annual tax amount
- monthly net salary

Assume 12 months.
Print everything nicely with f-strings.


````python
# Problem 4 – Simple Salary Breakdown

annual_gross_salary = float(input("Annual gross salary (€): "))
tax_rate_percent = float(input("Tax rate (%): "))

tax_rate = tax_rate_percent / 100

monthly_gross_salary = annual_gross_salary / 12
annual_tax_amount = annual_gross_salary * tax_rate
annual_net_salary = annual_gross_salary - annual_tax_amount
monthly_net_salary = annual_net_salary / 12

print(f"Monthly gross salary: €{monthly_gross_salary:.2f}")
print(f"Annual tax amount: €{annual_tax_amount:.2f}")
print(f"Monthly net salary: €{monthly_net_salary:.2f}")
````

## 5. Future Value from User-Friendly Input

Ask the user for:
- amount in euros
- interest rate in % (e.g. 5 for 5%)
- number of years

Convert correctly and calculate:


$FV = PV \times (1 + r)^{n}$


Then print:
```text
After N years, your money will be worth €X.XX

Note: Round to 2 decimal places

````python
# Problem 5 – Future Value from User-Friendly Input

amount = float(input("Amount in euros: "))
interest_rate_percent = float(input("Annual interest rate (in %): "))
years = int(input("Number of years: "))

interest_rate = interest_rate_percent / 100
future_value = amount * (1 + interest_rate) ** years

print(f"After {years} years, your money will be worth €{future_value:.2f}")
````

## 6. Invoice Formatter

You are given:
- product name (string)
- unit price (float)
- quantity (int)
- tax rate (%)

Compute:
1. subtotal
2. tax amount
3. total to pay

Print one line:
```text
You bought 3x "Notebook" for a total of €XX.XX (including tax).


````python
# Problem 6 – Invoice Formatter

product_name = input("Product name: ")
unit_price = float(input("Unit price (€): "))
quantity = int(input("Quantity: "))
tax_rate_percent = float(input("Tax rate (%): "))

tax_rate = tax_rate_percent / 100

subtotal = unit_price * quantity
tax_amount = subtotal * tax_rate
total_to_pay = subtotal + tax_amount

print(f'You bought {quantity}x "{product_name}" for a total of €{total_to_pay:.2f} (including tax).')
````

## 8. Average Price of 3 Stocks

Ask the user for the prices of 3 stocks.

Then:
- compute average
- find highest
- find lowest

Print the 3 results clearly.
Use `sum(...)`, `max(...)`, `min(...)`.


````python
# Problem 8 – Average Price of 3 Stocks

price_1 = float(input("Price of stock 1: "))
price_2 = float(input("Price of stock 2: "))
price_3 = float(input("Price of stock 3: "))

total = sum((price_1, price_2, price_3))
average_price = total / 3
highest_price = max(price_1, price_2, price_3)
lowest_price = min(price_1, price_2, price_3)

print(f"Average price: €{average_price:.2f}")
print(f"Highest price: €{highest_price:.2f}")
print(f"Lowest price: €{lowest_price:.2f}")
````

## 9. Purchasing Power After Inflation

Use:

amount = 1000
inflation_rate = 0.03   # 3% per year
years = 4

Calculate:
$real\_ value = \frac{amount}{(1+inflation\_ rate)^{years}}$

Then print:
€1000 today ≈ €... in 4 years with 3% inflation.

Round to 2 decimal points

````python
# Problem 9 – Purchasing Power After Inflation

amount = 1000
inflation_rate = 0.03  # 3% per year
years = 4

future_amount = amount / (1 + inflation_rate) ** years

print(f"€{amount} today ≈ €{future_amount:.2f} in {years} years with {inflation_rate * 100:.0f}% inflation.")
````

## 10. Portfolio Weights (no lists yet)

You invest in 3 companies:
- A: €400
- B: €350
- C: €250

1. Compute total invested  
2. Compute % weight of each company  
3. Print the 3 weights rounded to 2 decimals

Example:
```text
Total invested: €1000
A: 40.0%
B: 35.0%
C: 25.0%


````python
# Problem 10 – Portfolio Weights (no lists yet)

A = 400
B = 350
C = 250

total_invested = A + B + C

weight_A = (A / total_invested) * 100
weight_B = (B / total_invested) * 100
weight_C = (C / total_invested) * 100

print(f"Total invested: €{total_invested:.2f}")
print(f"A: {weight_A:.2f}%")
print(f"B: {weight_B:.2f}%")
print(f"C: {weight_C:.2f}%")
````
